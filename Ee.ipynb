```python
import pandas as pd
```

```python
df = pd.read_csv("Bedroom_Fan_1.csv")
```

```python
def assign_fan_speed(temp, hum):
    # Fan Speed 0 (Off)
    if temp == 19 and 89 <= hum <= 92:
        return 0
    elif temp == 20 and 64 <= hum <= 93:
        return 1

    # Fan Speed 1-3 (Low)
    elif 22 <= temp <= 25 and 58 <= hum <= 95:
        return 1
    elif 26 <= temp <= 27 and 83 <= hum <= 95:
        return 2
    elif 28 <= temp <= 30 and 29 <= hum <= 95:
        return 3

    # Fan Speed 4-6 (Medium)
    elif 31 <= temp <= 34 and 17 <= hum <= 95:
        return 4
    elif 35 <= temp <= 37 and 55 <= hum <= 86:
        return 5
    elif 38 <= temp <= 40 and 5 <= hum <= 60:
        return 6

    # Fan Speed 7-9 (Fastest)
    elif temp > 40:
        return 7
    
    else:
        return 8  # Default for any other range not covered

# Apply the function to create the fanSpeed column
df['fanSpeed'] = df.apply(lambda row: assign_fan_speed(row['temperature'], row['humidity']), axis=1)

```
